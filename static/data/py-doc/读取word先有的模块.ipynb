```python
from docx import Document
from docx.enum.text import WD_BREAK
# Document 类，不仅可以新建word文档，也可以打开一个本地文档
doc = Document('word/台州市气象局_一周天气_第1期.docx')  # 想获取的文档文件名，这里是相对路径。
for d in range(len(doc.paragraphs)):
    inline = doc.paragraphs[d].runs
    for i in inline:
        txt_initial = i.text
        xml = i._element.xml
        # print("找到分页符")
        print(xml)
        if 'w:br' in xml and 'type="page"' in xml:
            print("找到分页符")
            doc.paragraphs[d].add_run().add_break(docx.enum.text.WD_BREAK.PAGE)
```
