```python
from docx import Document
from docx.shared import Inches
from docx import Document
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import  Pt
from docx.oxml import OxmlElement
from docx.oxml.ns import  qn
from docx.shared import Inches
from docx.shared import Cm
from docx.text.run import Run
from docx.oxml.text.run import CT_R
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT, WD_LINE_SPACING

class paragraph_doc:
    def __init__(self,data):
        self.data = data
    def run(self):
        document = Document()
        # 文档标题
        head_1st = document.add_heading()     
        head_1st.paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER
        head_1st.paragraph_format.space_before = Pt(15)
        head_1st.paragraph_format.space_after = Pt(15)
        run = head_1st.add_run(self.data['title'][0],0)
        run.font.name = u'宋体' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(18)
        # 文档附标题
        head_sub = document.add_heading()     
        head_sub.paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT
        head_sub.paragraph_format.space_before = Pt(15)
        head_sub.paragraph_format.space_after = Pt(15)
        run = head_sub.add_run(self.data['subtitle'][0],0)
        run.font.name = u'宋体' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(16)
        # 加入段落
        p_main = document.add_paragraph()
        paragraph_format = p_main.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT  #段落居左
        paragraph_format.space_before = Pt(15)  #设置段落前间距
        paragraph_format.space_after = Pt(15)   #设置段落后间距
        paragraph_format.first_line_indent = Inches(0.3)  #第一行缩进
        paragraph_format.line_spacing_rule = WD_LINE_SPACING.ONE_POINT_FIVE # 1.5倍行距
        context_main = self.data['text']
        run = p_main.add_run(context_main)
        run.font.name = u'仿宋_GB2312' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'仿宋_GB2312')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(14)
        # 添加数据
        run = p_main.add_run("这里是测试文档")
        # paragraph_format = p_main.paragraph_format  #创建段落格式对象
        # paragraph_format.alignment = WD_ALIGN_PARAGRAPH.RIGHT  #段落居左
        run.font.name = u'仿宋_GB2312' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'仿宋_GB2312')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(14)

        
        
        # 保存文档
        document.save('word/测试.docx')
        pass












data = {
    "title":["本周后期强冷空气影响，全市有严重冰冻",1,2,3],
    "subtitle":["短信内容",1,2,3],
    "text":"本周前期天气阴沉本周后期强冷空气影响，全市有严重冰冻本周后期强冷空气影响，全市有严重冰冻本周后期强冷空气影响，全市有严重冰冻。"    
       }

a = paragraph_doc(data)
a.run()



```
