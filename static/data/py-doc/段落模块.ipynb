```python
from docx import Document
from docx.shared import Inches
from docx import Document
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import  Pt
from docx.oxml import OxmlElement
from docx.oxml.ns import  qn
from docx.shared import Inches
from docx.shared import Cm
from docx.text.run import Run
from docx.oxml.text.run import CT_R


class paragraph_doc:
    def __init__(self,data):
        self.data = data
    def run(self):
        document = Document()
        # 文档标题
        head_1st = document.add_heading()     
        head_1st.paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER
        head_1st.paragraph_format.space_before = Pt(0)
        head_1st.paragraph_format.space_after = Pt(0)
        run = head_1st.add_run(self.data['title'][0],0)
        run.font.size=Pt(25)
        run.font.color.rgb=RGBColor(0x00,0x00,0x00) #标题颜色
        # 文档附标题
        head_sub = document.add_heading()     
        head_sub.paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT
        head_sub.paragraph_format.space_before = Pt(0)
        head_sub.paragraph_format.space_after = Pt(0)
        run = head_sub.add_run(self.data['subtitle'][0],0)
        run.font.size=Pt(25)
        run.font.color.rgb=RGBColor(0x00,0x00,0x00) #标题颜色
        # 加入段落
        p_main = document.add_paragraph()
        paragraph_format = p_main.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT  #段落居左
        paragraph_format.space_before = Pt(0)  #设置段落前间距
        paragraph_format.space_after = Pt(0)   #设置段落后间距
        paragraph_format.first_line_indent = Inches(0.2)  #第一行缩进
        context_main = self.data['text']
        run = p_main.add_run(context_main)
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(10)
        
        # 保存文档
        document.save('word/段落.docx')
        pass












data = {
    "title":["我是谁",1,2,3],
    "subtitle":["我是你爹",1,2,3],
    "text":"党章第五条规定，发展党员必须严格履行以下手续：“申请入党的人，要填写入党志愿书，要有两名正式党员作介绍人。要经过支部大会通过和上级党组织批准，并且经过预备期的考察，才能成为正式党员。”根据这些规定，申请入党的人要经过以下主要程序才能成为正式党员： "
    
       }

a = paragraph_doc(data)
a.run()



```
