```python
from docx import Document
from docx.shared import Inches
from docx import Document
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import  Pt
from docx.oxml import OxmlElement
from docx.oxml.ns import  qn
from docx.shared import Inches
from docx.shared import Cm

from docx.text.run import Run
from docx.oxml.text.run import CT_R
from PIL import Image
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT, WD_LINE_SPACING
class foot_doc:
    def __init__(self,data):
        self.data = data
    def read_base64(self):
        pass
    def run(self):
        document = Document()
        # 保存
        png_obj = document.add_paragraph()
        paragraph_format = png_obj.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
        paragraph_format.space_before = Pt(0)  #设置段落前间距
        paragraph_format.space_after = Pt(0)   #设置段落后间距
        run1 = png_obj.add_run()
        run1.add_picture(self.data['png1'],width=Inches(1.5))
        png_obj.add_run("     ")
        run2 = png_obj.add_run()
        run2.add_picture(self.data['png2'],width=Inches(1.5))  
        png_obj.add_run("     ")
        run3 = png_obj.add_run()
        run3.add_picture(self.data['png3'],width=Inches(1.5)) 
        # 添加标题
        p_index = document.add_paragraph()
        paragraph_format = p_index.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
        paragraph_format.space_before = Pt(0)  #设置段落前间距
        paragraph_format.space_after = Pt(0)   #设置段落后间距
        blank_tab = "       "
        run = p_index.add_run(self.data['title1']+blank_tab+ self.data['title2'] + blank_tab + self.data['title3'] )
        run.font.name = u'宋体' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(14) 
        # 添加黑色实线
        document.add_picture('src/line-black.png')
        # 呈
        p_main = document.add_paragraph()
        paragraph_format = p_main.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT  #段落居左
        paragraph_format.space_before = Pt(0)  #设置段落前间距
        paragraph_format.space_after = Pt(0)   #设置段落后间距
        paragraph_format.first_line_indent = Inches(0)  #第一行缩进
        # paragraph_format.line_spacing_rule = WD_LINE_SPACING.ONE_POINT_FIVE # 1.5倍行距
        context_main = "呈:" + self.data['service_name']
        run = p_main.add_run(context_main)
        run.font.name = u'仿宋_GB2312' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'仿宋_GB2312')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(12.5)
        # 送
        p_main = document.add_paragraph()
        paragraph_format = p_main.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT  #段落居左
        paragraph_format.space_before = Pt(0)  #设置段落前间距
        paragraph_format.space_after = Pt(0)   #设置段落后间距
        paragraph_format.first_line_indent = Inches(0)  #第一行缩进
        # paragraph_format.line_spacing_rule = WD_LINE_SPACING.ONE_POINT_FIVE # 1.5倍行距
        context_main = "送:" + self.data['service_unity']
        run = p_main.add_run(context_main)
        run.font.name = u'仿宋_GB2312' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'仿宋_GB2312')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(12.5)
        # 发
        p_main = document.add_paragraph()
        paragraph_format = p_main.paragraph_format  #创建段落格式对象
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.LEFT  #段落居左
        paragraph_format.space_before = Pt(0)  #设置段落前间距
        paragraph_format.space_after = Pt(0)   #设置段落后间距
        paragraph_format.first_line_indent = Inches(0)  #第一行缩进
        # paragraph_format.line_spacing_rule = WD_LINE_SPACING.ONE_POINT_FIVE # 1.5倍行距
        context_main = "发:" + self.data['recive_unity']
        run = p_main.add_run(context_main)
        run.font.name = u'仿宋_GB2312' 
        run._element.rPr.rFonts.set(qn('w:eastAsia'), u'仿宋_GB2312')
        run.font.color.rgb = RGBColor(0, 0, 0)
        run.font.size = Pt(12.5)
        



        document.save('word/文件尾.docx')

        

data = {
    "type":"文件尾",
    "title1":"微信宣传",
    "png1":'src/二维码.png',
    "title2":"气象台宣传",
    "png2":'src/二维码.png',
    "title3":"浙政钉宣传",
    "png3":'src/二维码.png',
    "service_name":'宋伟2、李渊、小王、张三',
    "service_unity":'气象台、玉环市气象局、水利局、气象台、玉环市气象局、水利局、气象台、玉环市气象局、水利局、玉环市气象局、水利局、气象台、玉环市气象局、水利局、玉环市气象局、水利局、气象台、玉环市气象局、水利局、玉环市气象局、水利局、气象台、玉环市气象局、水利局',
    "recive_unity":'气象台、玉环市气象局、水利局、玉环市气象局、水利局'
       }

a = foot_doc(data)
a.run()
```
