```python
from docx import Document
from docx.shared import Inches
from docx import Document
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import  Pt
from docx.oxml import OxmlElement
from docx.oxml.ns import  qn
from docx.shared import Inches
from docx.shared import Cm

from docx.text.run import Run
from docx.oxml.text.run import CT_R
from PIL import Image
class png_doc:
    def __init__(self,data):
        self.data = data
    def read_base64(self):
        pass
    def run(self):
        document = Document()
        # 保存
        if self.data['type'] == '单图':
            document.add_picture(self.data['png'], width=Inches(5.8))
            # 居中的标题
            p_index = document.add_paragraph()
            paragraph_format = p_index.paragraph_format  #创建段落格式对象
            paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
            paragraph_format.space_before = Pt(0)  #设置段落前间距
            paragraph_format.space_after = Pt(0)   #设置段落后间距
            run = p_index.add_run("图:" + "  "+ str(self.data['title']))
            run.font.name = u'楷体' 
            run._element.rPr.rFonts.set(qn('w:eastAsia'), u'楷体')
            run.font.color.rgb = RGBColor(0, 0, 0)
            run.font.size = Pt(14)
        elif self.data['type'] == '双图':
            png_obj = document.add_paragraph()
            paragraph_format = png_obj.paragraph_format  #创建段落格式对象
            paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
            run1 = png_obj.add_run()
            run1.add_picture(self.data['png1'],width=Inches(2.7))
            run2 = png_obj.add_run()
            run2.add_picture(self.data['png1'],width=Inches(2.7))            
            # 居中的标题
            p_index = document.add_paragraph()
            paragraph_format = p_index.paragraph_format  #创建段落格式对象
            paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
            paragraph_format.space_before = Pt(0)  #设置段落前间距
            paragraph_format.space_after = Pt(0)   #设置段落后间距
            run = p_index.add_run("图1:" + "  "+ str(self.data['title1']) +"图2:" + "  "+ str(self.data['title2']) )
            run.font.name = u'楷体' 
            run._element.rPr.rFonts.set(qn('w:eastAsia'), u'楷体')
            run.font.color.rgb = RGBColor(0, 0, 0)
            run.font.size = Pt(14)
        else :
            png_obj1 = document.add_paragraph()
            paragraph_format = png_obj1.paragraph_format  #创建段落格式对象
            paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
            paragraph_format.space_before = Pt(0)  #设置段落前间距
            paragraph_format.space_after = Pt(0)   #设置段落后间距
            run1 = png_obj1.add_run()
            run1.add_picture(self.data['png1'],width=Inches(2.7))
            run2 = png_obj1.add_run()
            run2.add_picture(self.data['png1'],width=Inches(2.7))   

            png_obj2 = document.add_paragraph()
            paragraph_format = png_obj2.paragraph_format  #创建段落格式对象
            paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
            paragraph_format.space_before = Pt(0)  #设置段落前间距
            paragraph_format.space_after = Pt(0)   #设置段落后间距
            run1 = png_obj2.add_run()
            run1.add_picture(self.data['png3'],width=Inches(2.7))
            run2 = png_obj2.add_run()
            run2.add_picture(self.data['png4'],width=Inches(2.7))  

            # 居中的标题
            p_index = document.add_paragraph()
            paragraph_format = p_index.paragraph_format  #创建段落格式对象
            paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER   #段落居中对齐
            paragraph_format.space_before = Pt(0)  #设置段落前间距
            paragraph_format.space_after = Pt(0)   #设置段落后间距
            run = p_index.add_run("图1:" + "  "+ str(self.data['title1']) +"图2:" + "  "+ str(self.data['title2']) + \
                 "图3:" + "  "+ str(self.data['title3']) +"图4:" + "  "+ str(self.data['title4']) )
            run.font.name = u'楷体' 
            run._element.rPr.rFonts.set(qn('w:eastAsia'), u'楷体')
            run.font.color.rgb = RGBColor(0, 0, 0)
            run.font.size = Pt(12.5)
        document.save('word/图片.docx')

        
##  数据的基本格式    
# data = {
#     "type":"单图",
#     "title":"台州市气象台",
#     "png":'src/pre_test.png'
#        }
# data = {
#     "type":"双图",
#     "title1":"台州市气象台1",
#     "png1":'src/pre_test1.png',
#     "title2":"台州市气象台2",
#     "png2":'src/pre_test2.png'
#        }
data = {
    "type":"四图",
    "title1":"台州市气象台1",
    "png1":'src/pre_test1.png',
    "title2":"台州市气象台2",
    "png2":'src/pre_test2.png',
    "title3":"台州市气象台3",
    "png3":'src/pre_test1.png',
    "title4":"台州市气象台4",
    "png4":'src/pre_test1.png'
       }

a = png_doc(data)
a.run()
```
